```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the AP dataset
AP = pd.read_csv("2012__AP_Results_20241118.csv")

# Load the SAT dataset
SAT = pd.read_csv("2012_SAT_Results_20241118.csv")

print("AP Dataset Sample:")
print(AP.head())

print("\nSAT Dataset Sample:")
print(SAT.head())
```

```python
print("AP Dataset Columns:")
print(AP.columns)

print("\nSAT Dataset Columns:")
print(SAT.columns)
```

```python
AP["SCHOOL NAME"] = AP["SCHOOL NAME"].str.strip().str.upper()
SAT["SCHOOL NAME"] = SAT["SCHOOL NAME"].str.strip().str.upper()
```

```python
merged_data = pd.merge(AP, SAT, on="SCHOOL NAME", how="inner")
score_columns = [
    "SAT Critical Reading Avg. Score",
    "SAT Math Avg. Score",
    "SAT Writing Avg. Score",
]

for col in score_columns:
    merged_data[col] = pd.to_numeric(merged_data[col], errors="coerce")
```

```python
# Drop rows where all SAT score columns are NaN
merged_data.dropna(subset=score_columns, how="all", inplace=True)
```

```python
merged_data["Avg SAT Score"] = merged_data[score_columns].mean(axis=1)
```

```python
print("\nMerged Dataset Sample:")
print(merged_data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure SAT score columns and AP columns are numeric
score_columns = [
    "SAT Critical Reading Avg. Score",
    "SAT Math Avg. Score",
    "SAT Writing Avg. Score"
]
merged_data[score_columns] = merged_data[score_columns].apply(pd.to_numeric, errors="coerce")
merged_data["Num of AP Exams Passed"] = pd.to_numeric(merged_data["Num of AP Exams Passed"], errors="coerce")
merged_data["Num of AP Test Takers"] = pd.to_numeric(merged_data["Num of AP Test Takers"], errors="coerce")

# Calculate the total SAT score
merged_data["Total SAT Score"] = merged_data[score_columns].sum(axis=1)

# Simplify the legend by binning 'AP Exams Passed' into ranges
bins = [0, 10, 50, 100, 500, 1000, 5000, 10000]
labels = ["0-10", "11-50", "51-100", "101-500", "501-1000", "1001-5000", "5000+"]
merged_data["AP Exams Passed (Binned)"] = pd.cut(merged_data["Num of AP Exams Passed"], bins=bins, labels=labels)

# Drop rows where Total SAT Score or AP data is NaN
merged_data = merged_data.dropna(subset=["Total SAT Score", "Num of AP Test Takers", "Num of AP Exams Passed"])

# Plot the relationship between Total SAT Score and AP Test Participation
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=merged_data,
    x="Total SAT Score",
    y="Num of AP Test Takers",
    hue="AP Exams Passed (Binned)",
    palette="viridis"
)
plt.title("Relationship Between Total SAT Scores and AP Test Participation")
plt.xlabel("Total SAT Score")
plt.ylabel("Number of AP Test Takers")
plt.legend(title="AP Exams Passed (Ranges)", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Analysis of the Graph and Data
print("Analysis:")
print("1. Most schools have Total SAT Scores in the range of 1000 to 1600, with fewer than 500 AP test takers.")
print("2. Schools with lower AP exams passed (bins like '0-10' or '11-50') are distributed across the full SAT score range.")
print("3. Schools in the higher bins (e.g., '501-1000', '1001-5000') tend to cluster at higher AP test taker numbers, suggesting a possible correlation.")
print("4. Outliers exist where schools have a disproportionately large number of AP test takers (e.g., above 2000) and high AP exams passed.")

# Calculate and display the correlation
correlation = merged_data["Total SAT Score"].corr(merged_data["Num of AP Test Takers"])
print("\nCorrelation:")
print(f"The correlation between Total SAT Scores and Number of AP Test Takers is: {correlation:.2f}")

# Conclusion
print("\nConclusion:")
print("1. Schools with higher total SAT scores tend to have more AP test takers and higher AP exam pass rates.")
print("2. However, the relationship is modest, as indicated by a correlation of {:.2f}.".format(correlation))
print("3. There is significant variation in the middle SAT score ranges, with some schools having high AP participation despite lower SAT scores.")
print("4. This suggests that other factors, such as school size or student encouragement, may influence AP participation alongside SAT performance.")
```
