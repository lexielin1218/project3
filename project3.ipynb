```python
# First thing, how many parks are there?

import pandas as pd

Park = pd.read_csv("park-data.csv", encoding="ISO-8859-1")

parks_by_borough = Park.groupby('Area Name')['Park Name'].unique().reset_index()

parks_by_borough.columns = ['Borough', 'Park Names']

print(parks_by_borough)
```

```python
# Second thing, how many squirrels are there?
import pandas as pd

# Load the dataset
Squirrel = pd.read_csv("squirrel-data.csv", encoding="ISO-8859-1")

# Group by 'Park Name' and count the squirrels
squirrels_per_park = Squirrel.groupby('Park Name').size().reset_index(name='Squirrel Count')

# Display the result
print(squirrels_per_park)
```

```python
# Next question, Which park has the highest squirrel population?
import matplotlib.pyplot as plt

top_squirrel_park = squirrels_per_park.sort_values(by='Squirrel Count', ascending=False).iloc[0]

squirrels_per_park.sort_values(by='Squirrel Count', ascending=False, inplace=True)
plt.figure(figsize=(12, 6))
plt.bar(squirrels_per_park['Park Name'], squirrels_per_park['Squirrel Count'])
plt.xticks(rotation=90)
plt.title('Number of Squirrels per Park')
plt.xlabel('Park Name')
plt.ylabel('Squirrel Count')
plt.show()

print(f"Park with the highest squirrel population: {top_squirrel_park['Park Name']} ({top_squirrel_park['Squirrel Count']} squirrels)")
```

```python
# Follow up question, What are the most common squirrel activities in each park?

activities_by_park = Squirrel.groupby(['Park Name', 'Activities']).size().reset_index(name='Count')

top_activities_by_park = activities_by_park.sort_values(by=['Park Name', 'Count'], ascending=[True, False])

print(top_activities_by_park)

most_populated_park = top_squirrel_park['Park Name']
activities_in_top_park = activities_by_park[activities_by_park['Park Name'] == most_populated_park]
plt.figure(figsize=(8, 5))
plt.bar(activities_in_top_park['Activities'], activities_in_top_park['Count'])
plt.title(f"Top Activities in {most_populated_park}")
plt.xlabel('Activities')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Some random questions...  Which park has the most active squirrels?

active_squirrels = Squirrel.groupby('Park Name')['Activities'].apply(lambda x: x.str.contains('Running|Foraging|Digging', na=False).sum()).reset_index(name='Active Squirrels')

most_active_park = active_squirrels.loc[active_squirrels['Active Squirrels'].idxmax()]
print(f"The park with the most active squirrels is {most_active_park['Park Name']} with {most_active_park['Active Squirrels']} active squirrels.")

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(active_squirrels['Park Name'], active_squirrels['Active Squirrels'])
plt.xticks(rotation=90)
plt.title('Active Squirrels per Park')
plt.xlabel('Park Name')
plt.ylabel('Number of Active Squirrels')
plt.show()
```

```python
#Are there more squirrels observed in busier parks (based on park conditions)?
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
Park = pd.read_csv("park-data.csv", encoding="ISO-8859-1")
Squirrel = pd.read_csv("squirrel-data.csv", encoding="ISO-8859-1")

# Calculate the number of squirrels in each park
squirrels_per_park = Squirrel.groupby('Park Name').size().reset_index(name='Squirrel Count')

# Merge park data with squirrel counts
merged_data = Park.merge(squirrels_per_park, on='Park Name', how='inner')

# Group by park conditions and calculate the average number of squirrels
squirrels_by_condition = merged_data.groupby('Park Conditions')['Squirrel Count'].mean().reset_index()

# Clean Park Conditions column (replace unsupported characters)
squirrels_by_condition['Park Conditions'] = squirrels_by_condition['Park Conditions'].str.encode('ascii', errors='ignore').str.decode('ascii')

# Plot average squirrel counts for each condition
plt.figure(figsize=(10, 6))
plt.bar(squirrels_by_condition['Park Conditions'], squirrels_by_condition['Squirrel Count'], color='orange')
plt.title('Average Squirrel Counts by Park Conditions')
plt.xlabel('Park Conditions')
plt.ylabel('Average Squirrel Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Finally, plotting the squirrel on the map!
import folium

cleaned_squirrel_data = Squirrel.dropna(subset=['Squirrel Latitude (DD.DDDDDD)', 'Squirrel Longitude (-DD.DDDDDD)'])

squirrel_map = folium.Map(location=[40.7831, -73.9712], zoom_start=12)

for _, row in cleaned_squirrel_data.iterrows():
    folium.CircleMarker(
        location=[row['Squirrel Latitude (DD.DDDDDD)'], row['Squirrel Longitude (-DD.DDDDDD)']],
        radius=3,
        color='orange',
        fill=True,
        fill_opacity=0.6
    ).add_to(squirrel_map)

squirrel_map
```
