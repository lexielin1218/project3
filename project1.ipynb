```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Buildings_by_Borough.csv")
df

# Extract the "Community District" column
community_district = df["Community District"]

# Calculate mean, median, and mode for the "Community District" column
mean_cd = community_district.mean()
median_cd = community_district.median()
mode_cd = community_district.mode().iloc[0]  # taking the first mode if there are multiple

# Print the results
print("Mean:", mean_cd)
print("Median:", median_cd)
print("Mode:", mode_cd)
```

```python
import csv
from collections import Counter

# Load the data
with open("Buildings_by_Borough.csv", mode="r") as file:
    reader = csv.DictReader(file)
    community_districts = [float(row["Community District"]) for row in reader if row["Community District"]]

# Calculate the mean
mean_cd = sum(community_districts) / len(community_districts)

# Calculate the median
sorted_cd = sorted(community_districts)
n = len(sorted_cd)
if n % 2 == 1:
    median_cd = sorted_cd[n // 2]
else:
    median_cd = (sorted_cd[n // 2 - 1] + sorted_cd[n // 2]) / 2

# Calculate the mode
count_cd = Counter(sorted_cd)
mode_cd = max(count_cd, key=count_cd.get)

# Print the results
print("Mean:", mean_cd)
print("Median:", median_cd)
print("Mode:", mode_cd)
```

```python
import csv
from collections import defaultdict

# Load data and compute the average "Community District" value per borough
with open("Buildings_by_Borough.csv", mode="r") as file:
    reader = csv.DictReader(file)
    borough_data = defaultdict(list)
    for row in reader:
        if row["Community District"]:
            borough = row["Borough"]
            community_district = float(row["Community District"])
            borough_data[borough].append(community_district)

# Calculate the average "Community District" for each borough
average_districts = {borough: sum(values) / len(values) for borough, values in borough_data.items()}

# Find the max average to scale the bars
max_average = max(average_districts.values())
scale_factor = 30 / max_average  # Adjusted scale factor to reduce bar length

# Generate ASCII bar chart
output = "\nAverage Community District by Borough\n" + "=" * 40 + "\n"
for borough, avg in average_districts.items():
    bar_length = int(avg * scale_factor)
    output += f"{borough}: {'*' * bar_length} ({avg:.2f})\n"

# Print the output
print(output)
```
